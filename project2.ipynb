 ## Section 4: Data Science: Analysis about Heart disease

 Now, we need to build a new dataframe called "df_HD" including all rows that the Cause name is Heart disease.


```python

dfsum = df[(df["Cause Name"] == "All causes")&(df["State"] == "United States")]
dfsum = dfsum.sort_values(by="Year",ascending= False)
dfsum
```

```python
df_H = df[(df["Cause Name"] == "Heart disease")&(df["State"] == "United States")]
df_H = df_H.sort_values(by="Year",ascending= False)
df_H
```

```python
dfsum.loc[:, 'Deaths']
tt = [2813503,2744248,2712630,2626418,2596993,2543279,2515458,2468435,2423712,2437163,2471984,2426264,2448017,2397615,2448288,2443387,2416425,2403351,2391399]
df_H["total"] = tt
df_H
```

 ### 4.1 The relationship between year and deaths

In order to have a correct understanding about relationship between year and deaths, we should notice that with the **growth of population**, the deaths will also increase. We need to try to diminish the influence of population growth.

Add a new column called "Ratio"

```python

df_H["Ratio"] = df_H["Deaths"]/df_H["total"]
df_H

```

```python
numerical_cols = ["Year","Ratio","Age-adjusted Death Rate"]
numerical_cols
```

```python
df_numerical = df_H[numerical_cols]
df_numerical
```

```python
df_numerical.corr()
```

```python
df_numerical.plot.scatter(x='Year',y='Ratio')
```

#### Analysis according to correltion matrix and linear correlation graph

As graph shows, "Year" and ratio that heart diseases occupies in total deaths have a strong and negative linear relationship which is -0.966732. The ratio of heart disease genenrally decreases when year becomes bigger. This probably means that with the development of modern medicine, there are more solutions to help people get appropriate treatment. In 19 years, the ratio declines almost 10%. However, this also means that some leading causes probably increases during this period.

 ### 4.2 Regression

```python
heart_disease = "Deaths"
heart_disease
```

```python
heart_disease_mean = df_H[heart_disease].mean()
heart_disease_mean
```

```python
heart_disease_std = df_H[heart_disease].std()
heart_disease_std
```

```python
heart_disease_size = df_H[heart_disease].count()
heart_disease_size
```

```python
import scipy.stats
standard_error = heart_disease_std/heart_disease_size**(1/2)
standard_error
```

```python
slope = (-0.966732)*(df_numerical["Ratio"].std())/(df_numerical["Year"].std())
slope
```

```python
intercept = df_numerical["Ratio"].mean()-slope*df_numerical["Year"].mean()
intercept
```

```python
import seaborn as sns
sns.regplot(data=df_H, x="Year", y="Ratio", ci=None, line_kws={'color':'red'})
```

```python
df_H['HD_prediction'] = (slope)*df_numerical["Year"] + intercept
df_H
```

```python
df_H['residual'] = df_numerical["Ratio"] - df_H['HD_prediction']
df_H
```

```python
RMSE = (1-(-0.966732)**2)**(1/2)
RMSE
```

I have built a regression line between year and ratio. The SD of the prediction errors is 0.256 which probably indicates that the prediction is generally accurate. However, I have noticed that the sample size is not big enough which is only 19. In order to make the prediction more accurate, there should be more data.
